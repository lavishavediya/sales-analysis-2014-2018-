### merge sales with customer information

```python
orders
```

```python
customers
```

```python
customers.rename(columns={"customer_id":"Customer Index"},inplace=True)
```

```python
orders.rename(columns={
    "customer_name_index":"Customer Index",
    "delivery_region_index" : "Region Index",
     "product_description_index":"Product Index"},inplace = True)
```

```python
merged_df = orders.merge(customers, on="Customer Index", how="left") 
```

```python
products.rename(columns={"Index":"product Index"},inplace= True)
```

```python
merged_df = merged_df.merge(products, left_on="Product Index", right_on="product Index", how="left")
merged_df.drop(columns=["product Index"],inplace=True)
```

```python
products.columns.tolist()
```

```python
merged_df=merged_df.merge(regions, left_on="Region Index", right_on="id",how="left")
merged_df.drop(columns=["id"],inplace=True)
```

```python
merged_df["Profit"]=merged_df["line_total"] - merged_df["total_unit_cost"]
merged_df["Profit Margin %"]=round((merged_df["Profit"]/merged_df["line_total"])*100,2)

```

```python
merged_df.head()
```

```python
merged_df.to_csv("cleaned_sales_data.csv",index=False)
```

```python
df = pd.read_csv("cleaned_sales_data.csv")
```

```python
columns_to_drop=['Customer Index', 'Product Index','Region Index','lattitude','longitutde','area_code', 'time_zone', 'land_area', 'water_area'
]
df=df.drop(columns=columns_to_drop,errors='ignore')
```

```python
df = df.rename(columns={
    'ordernumber': 'Order Number',
    'orderdate': 'Order Date',
    'channel': 'Sales Channel',
    'currency_code': 'Currency',
    'warehouse_code': 'Warehouse',
    'order_quantity': 'Order Quantity',
    'unit_price': 'Unit Price',
    'total_unit_cost': 'Total Unit Cost',
    'line_total': 'Line Total',
    'Profit': 'Profit',
    'Profit Margin %': 'Profit Margin (%)',
    'productname': 'Product Name',
    'category': 'Category',
    'subcategory': 'Subcategory',
    'brand': 'Brand',
    'customername': 'Customer Name',
    'regionname': 'Region Name'
})
```

```python
#converting order date and sorting
df["Order Date"] = pd.to_datetime(df["Order Date"], dayfirst=True, errors='coerce')
df = df.sort_values("Order Date")
```

```python
df.to_csv("final_cleaned_sales_data.csv",index=False)
```
